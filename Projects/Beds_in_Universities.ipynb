### Step1.導入pandas模組

````python
import pandas as pd
````

### Step2.讀取csv檔

````python
#109年資料
data=pd.read_csv('https://stats.moe.gov.tw/files/detail/109/109_land.csv')
data
````

### Step3.將多餘的欄位刪除

````python
data1 = data.drop(
    columns=['縣市別', '學校代碼', '教職員宿舍單身床位_床', '教職員宿舍有眷戶數_戶'], axis=1)
display(data1) #data1為data操作後的第一個版本故為1
````

### Step.4.分組
將相同大學名稱的欄位資訊相加

````python
group = data1.groupby(['學校名稱'])
sum = data1.groupby(['學校名稱'])['學生宿舍床位數_床',	'自有宿舍床位數_床', '可使用校地-校舍基地_平方公尺',	'可使用校地-露天運動場地_平方公尺',	'可使用校地-其他校地_平方公尺',
      	'無法使用校地_平方公尺'].agg('sum')
display(sum) #先用學校名稱作分組，然後再衣床位及校舍面積各欄位作相加總合用.agg()函數。
````

### A.1(a)計算Data中的學校數目

### A.1(b)計算Data中的各類別學校數目

### A.3(c)床位平均數量

````python
#color
total=0 #計數器初始值為零
for i in range(len(sum)):
    total=total+sum.loc[sum.index[i],'學生宿舍床位數_床'] #將全部的床位數加起來
    average=round(total/len(sum))#取平均
print(f"各類學校中共有\033[34m{total}\033[00m個床位\n")
print(f"平均每間學校為\033[34m{average}\033[00m床\n")
print(f"國立東華大學比平均數多\033[34m{sum.loc['國立東華大學','學生宿舍床位數_床']-average}\033[00m床")
````

### 可用校地面積數據分析

關於校地欄位單獨抓出來

````python
for i in range(len(sum)):
    can_use=pd.DataFrame()#建立一個空的DataFrame
    #將sum的欄位添加到這新的DataFrame中
    can_use['可使用校地-校舍基地_平方公尺']=sum.loc[:,'可使用校地-校舍基地_平方公尺']
    can_use['可使用校地-露天運動場地_平方公尺']=sum.loc[:,'可使用校地-露天運動場地_平方公尺']
    can_use['可使用校地-其他校地_平方公尺']=sum.loc[:,'可使用校地-其他校地_平方公尺']
    #這是新的欄位自己家的是將前三項作一個總和
    can_use['可使用校地-總和_平方公尺']=sum.loc[:,'可使用校地-校舍基地_平方公尺']+sum.loc[:,'可使用校地-露天運動場地_平方公尺']+sum.loc[:,'可使用校地-其他校地_平方公尺']
    can_use['無法使用校地_平方公尺']=sum.loc[:,'無法使用校地_平方公尺']
#display(can_use)
````

````python
#非科技大學的部分
max=0
for i in range(len(sum)):
    if (r"科技大學") not in sum.index[i]: #關鍵字是不要有科技大學這四個字
        if can_use.iloc[i, 3] > max: #和剛剛取最大值的操作是一樣的
            max = can_use.iloc[i, 3]
            total_area_general = max
            J = i #最大值的索引值
#因為上面算出來是東華大學故國立一般大學就找出來了

#學院的部分
max = 0
for i in range(len(sum)):
    if (r"學院")  in sum.index[i] :
        if can_use.iloc[i, 3] > max:
            max = can_use.iloc[i, 3]
            total_area_college = max
            H = i

#專科學校的部分
max=0
for i in range(len(sum)):
    if (r"專科學校") in sum.index[i]:
        if can_use.iloc[i, 3] > max:
            max = can_use.iloc[i, 3]
            total_area_Jcollege = max
            V = i

#私立學校的部分
max=0
for i in range(len(sum)):
    if(r'國立') not in sum.index[i]:
        if can_use.iloc[i, 3] > max:
            max = can_use.iloc[i, 3]
            total_area_private = max
            K = i

#輸出結果
print(f"\033[92m{can_use.index[J]}\033[00m為一般大學中可用校地最大者,有\033[34m{total_area_general}\033[00m平方公尺", "\n")
print(f"\033[92m{can_use.index[H]}\033[00m為學院中可用校地最大者,有\033[34m{total_area_college}\033[00m平方公尺", "\n")
print(f"\033[92m{can_use.index[V]}\033[00m為專科學校中可用校地最大者,有\033[34m{total_area_Jcollege}\033[00m平方公尺", "\n")
print(f"\033[92m{can_use.index[K]}\033[00m為私立學校中可用校地最大者,有\033[34m{total_area_private}\033[00m平方公尺","\n")
````

### 6.歷年資料比對
讀取歷年csv檔

````python
#105-108的csv檔讀取
data_105=pd.read_csv('https://stats.moe.gov.tw/files/detail/105/105_land.csv')
data_106 = pd.read_csv('https://stats.moe.gov.tw/files/detail/106/106_land.csv')
data_107 = pd.read_csv('https://stats.moe.gov.tw/files/detail/107/107_land.csv')
data_108=pd.read_csv('https://stats.moe.gov.tw/files/detail/108/108_land.csv')

#將他們作成列表
data_list=[data_105,data_106,data_107,data_108]
````

將歷年的資料做刪去欄位及相同學校相同欄位的資訊相加

````python
old_sum_list=[] #建立一個空列表用來承接將不要的欄位刪除後的歷年DataFrame
for i in range(0,4):#list只有4個東西0-3
    year=105+i #學年
    data_list[i].drop(
        columns=['縣市別', '學校代碼', '教職員宿舍單身床位\n(床)', '教職員宿舍有眷戶數\n(戶)'], axis=1)
    #將不要的欄位做刪除

    group = data_list[i].groupby(['學校名稱'])
    #在依學校名稱做分類
    old_sum_list.append(data_list[i].groupby(['學校名稱'])['學生宿舍床位數\n(床)','可使用校地-校舍基地(平方公尺)', '可使用校地-露天運動場地\n(平方公尺)', '可使用校地-其他校地\n(平方公尺)', '無法使用校地\n(平方公尺)'].agg('sum'))
    #再將相同欄位及學校相同的資訊做相加

    #display(f"{105+i}年",old_sum_list[i])

    N = 0#計算歷年學校數量的消長
    for school in group:
        N = N+1
    print(f"{year}年的data中總共有\033[34m{N}\033[00m間學校\n")
````

### A.6(a)105-108歷年床位排名

````python
bed_total=[]#這個空列表是用來裝歷年床位總和值的列表
for i in range(len(old_sum_list)):#0-3 只是用len的形式表達
    bed = old_sum_list[i].rank().sort_values(by='學生宿舍床位數\n(床)')
    #歷年都做
    #一樣先做rank做排序

    #display(bed)

    #輸出前10
    print(f"\n{105+i}年床位最多的前十名排行\n")
    for rank in range(1, 11):
        print(f"第\033[93m{rank}\033[00m名為", bed.index[-rank], ",有", " \033[34m",old_sum_list[i].loc[bed.index[-rank], '學生宿舍床位數\n(床)'], " \033[00m", "床", "\n")

    print("\n")

    #輸出最少前15
    print(f"{105+i}年床位最少的前十五名排行\n")
    for rank in range(0, 15):
        print(f"第\033[93m{rank+1}\033[00m名為", bed.index[rank], ", 有", "\033[93m", old_sum_list[i].loc[bed.index[rank], '學生宿舍床位數\n(床)'],"\033[00m", "床", "\n")

    #這裡是將年床位數做一個總和
    total=0# 初始值為零
    for h in range(len(old_sum_list[i])): #次數為該年次的列數
        total=total+old_sum_list[i].iloc[h, 0]#將所有列做相加
    bed_total.append(total)#計算好後裝到表中
````

### A.6(b)105-108歷年學校可用土地排名

````python
total_area=[]#空列表用來裝計算後的校地總和的值

for j in range(len(old_sum_list)):#0-3 四年份
    for i in range(len(old_sum_list[j])):#每一年的學校數量皆不一樣
        #建立空的DataFrame
        old_can_use = pd.DataFrame()
 # 將欄位為以下的欄放入DataFrame中
        old_can_use['可使用校地-校舍基地(平方公尺)'] = old_sum_list[j].loc[: , '可使用校地-校舍基地(平方公尺)']
        old_can_use['可使用校地-露天運動場地\n(平方公尺)'] = old_sum_list[j].loc[:, '可使用校地-露天運動場地\n(平方公尺)']
        old_can_use['可使用校地-其他校地\n(平方公尺)'] = old_sum_list[j].loc[:, '可使用校地-其他校地\n(平方公尺)']
        old_can_use['可使用校地-總和(平方公尺)'] = old_sum_list[j].loc[:, '可使用校地-校舍基地(平方公尺)'] + old_sum_list[j].loc[:, '可使用校地-露天運動場地\n(平方公尺)']+old_sum_list[j].loc[:, '可使用校地-其他校地\n(平方公尺)']
        old_can_use['無法使用校地\n(平方公尺)'] = old_sum_list[j].loc[:, '無法使用校地\n(平方公尺)']


    old_can_use_rank = old_can_use.rank(method='average').sort_values(by='可使用校地-總和(平方公尺)')
    #先做rank做排序

    #display(old_can_use_rank)

    #排行榜
    
    print("\n",105+j,"年全國可用校地面積排名\n")#105+j為年分
  
    for rank in range(1, 11):
        print(f"第\033[93m{rank}\033[00m名為\033[92m{old_can_use_rank.index[-rank]}\033[00m")
        print("可使用面積為\033[34m", old_can_use.loc[old_can_use_rank.index[-rank],'可使用校地-總和(平方公尺)'], "\033[00m平方公尺\n") 

    #求歷年可用校地總和
    total = 0  # 歷年可用土地總和初始值為零
    for v in range(len(old_sum_list[j])):  # 次數為該年次的列數
        total = total+old_can_use.iloc[v, 3]  # 將所有列做相加
    total_area.append(total)  # 計算好後裝到表中
#display(old_can_use)
````
